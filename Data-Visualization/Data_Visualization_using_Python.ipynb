```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_color_codes('dark')
%matplotlib inline 
# output should be within the notebook
```

```python
auto= pd.read_csv("Automobile.csv")
auto
```

```python
sns.distplot(auto['normalized_losses'], rug=True)
```

```python
sns.distplot(auto['wheel_base'], kde=False,rug=True,)
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'])
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'],kind='reg',
    color='black',)
```

```python
sns.pairplot(auto[['engine_size','horsepower','width']])
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'],hue= auto['fuel_type'])
```

```python
sns.countplot(auto['body_style'], hue= auto['engine_location'])
```
